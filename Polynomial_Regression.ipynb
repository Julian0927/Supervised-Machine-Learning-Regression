# ***Polynomial Regression***



What if your data is more complex than a straight line? Surprisingly, you can use a linear model to fit nonlinear data. A simple way to do this is to add powers of each feature as new features, then train a linear model on this extended set of features. This technique is called $\color{aquamarine}{\text{Polynomial Regression}}$.

$ \ $

There are two factors when determining model performance: 

$(1)$ $\color{yellow}{\text{Overfitting}}$.

$\color{yellow}{\text{Overfitting}}$ is when the model is too complex and does well on the training data but not on the test data. $\color{yellow}{\text{Overfitting}}$ is simple to deal with, using methods like regularization, which we will discuss in the next lab.

$ \ $

$(2)$ $\color{ORANGE}{\text{Underfitting}}$. 

$\color{ORANGE}{\text{Underfitting}}$ is when the model is too simple and performs poorly on the training and testing data sets. To deal with $\color{ORANGE}{\text{underfitting}}$, we can build a more complex model using methods like polynomial regression. If making a more complex model does not work, this may involve using more data to train the model on or obtaining new features. As this process is complex, it's better to determine if the model can overfit the data first. Therefore, in this section, we will use Polynomial Regression to overfit the data to determine if we have an adequate amount of data.

## ***GridSearch and  Pipeline***


In this section of the notebook, we will define a pipeline object, then use $\color{aquamarine}{\text{GridSearch}}$ to find the best hyper-parameters of the model by using  $\color{orange}{\text{cross-validation}}$ method of the  parameter grid, as shown in the figure below. A 5-fold cross validation is used by default. We will learn more about k-fold cross validation in the next, Cross Validation lesson of the Course.

$ \ $

$(1)$ We create `PolynomialFeatures()` pipeline.


```python
# creamos una lista con los modelos PolynomialFeatures y LinearRegression
Input = [ ('polynomial', PolynomialFeatures(degree = 2, include_bias = False) ), ('model', LinearRegression() ) ]

# creamos una tuberia con los modelos anteriores
pipe = Pipeline(Input)
```

$ \ $

$(2)$ To search for the best combination of hyperparameters, we create a  <b>GridSearchCV</b> object with dictionary of parameter values. The parameters of pipelines can be set using the name of the key separated by '\__', then the parameter name of the key.


* <b>polynomial\_\_degree</b>: is the degree of the polynomial.

* <b>model\_\_normalize</b> This parameter is ignored when fit_intercept is set to False. If True, the regressors $X$ will be normalized before regression by subtracting the mean and dividing by the $L^{2}$-norm.



```python
param_grid = {"polynomial__degree": [1, 2, 3], "model__normalize": [True, False]}
```

$ \ $

$(3)$ The model is overfitting.


```python
search = GridSearchCV(pipe, param_grid, n_jobs = 1)
```

```python
# entrenamos nuestra tuberia con la data X_train, y_train
pipe.fit(X_train, y_train)
```

```python
search.fit(X_test, y_test)
```

$ \ $

$(4)$ The object finds the best parameter values on the validation data. We can obtain the estimator with the best parameters and assign it to the variable `best`, as follows:


```python
best=search.best_estimator_
best
```

We see the degree is one and normalize is <code>Ture </code>.


$ \ $

$(5)$ Now, we test our model on the test data to see the best score.


```python
best.score(X_test,y_test)
```

$ \ $

$(6)$ Finally, we can plot a distribution of the predicted values versus the actual values.


```python
predicted=best.predict(X_test)
plot_dis(y_test,predicted)
```

As we see, the result is almost perfect!

```python
# creamos una lista con la data X
features = list(X)

# creamos dos listas adicionalmente donde guardaremos nuestra data
R_2_train=[]
R_2_test=[]

# realizamos una iteracion sobre features
for feature in features:
    # creamos un diccionario con estos datos
    param_grid = {"polynomial__degree": [ 1,2,3,4,5], "model__positive":[True, False]}

    # creamos una lista con los dos modelos a usar los cuales son PolynomialFeatures y LinearRegression
    Input=[ ('polynomial', PolynomialFeatures(degree = 2, include_bias = False) ), ('model', LinearRegression() )]

    # creamos una tuberia con los modelos anteriores
    pipe = Pipeline(Input)

    # mostramos en pantalla el rasgo de la iteracion 
    print(feature)

    # usamos la funci√≥n GridSearchCV sobre la tuberia creada anteriormente
    search = GridSearchCV(pipe, param_grid, n_jobs = 2)

    # entrenamos el modelo search con la data X_test[[feature]], y_test
    search.fit(X_test[[feature]], y_test)

    # usamos esta funcion
    best = search.best_estimator_

    # pegamos en nuestra lista best.score(X_test[[feature]], y_test)
    R_2_test.append(best.score(X_test[[feature]], y_test))

    # pegamos en nuestra lista best.score(X_train[[feature]], y_train)
    R_2_train.append(best.score(X_train[[feature]], y_train))

##################### todo esto es para obtener las listas R_2_train y R_2_test


# creamos un grafico de barras con la data features y  R_2_train
plt.bar(features, R_2_train, label = "Train")

# creamos un grafico de barras  con la data features, R_2_test
plt.bar(features, R_2_test, label = "Test")

# los datos del eje x los rotamos 90 grados
plt.xticks(rotation = 90)

# mostramos las etiquetas del eje y
plt.ylabel("$R^2$")

# mostramos la leyenda del plot
plt.legend()

# mostramos el plot en pantalla
plt.show()

# mostramos en pantalla los resultados del promedio de R^2 y el valor maximo
print(f"Training R^2 mean value {str(np.mean(R_2_train))} Testing R^2 mean value {str(np.mean(R_2_test))}")
print(f"Training R^2 max value {str(np.max(R_2_train))} Testing R^2 max value {str(np.max(R_2_test))} ")
```
