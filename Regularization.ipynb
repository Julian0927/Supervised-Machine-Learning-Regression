## ***Useful plotting functions***

In the following cells we load the data and define some useful plotting functions.


```python
from sklearn.datasets import fetch_california_housing

housing = fetch_california_housing()
```

```python
np.random.seed(72018)

def to_2d(array):
    return array.reshape(array.shape[0], -1)
```

```python
def california_dataframe(description = False):

    #(1) buscar vivivienda en california
    california = fetch_california_housing()
    
    #(2) cargamos los rasgos en las correspondientes variables
    data = california["data"]
    target = california["target"]
    names = california["feature_names"]
    
    #(3) usamos la funcion to_2d
    target = to_2d(target)
    
    #(4) concatenamos algunos datos
    data_all = np.concatenate([data, target], axis=1)
    names_all = np.concatenate([names, np.array(['MEDV'])], axis=0)
    
    #(5) 
    if description:
        
        return pd.DataFrame(data = data_all, columns = names_all), california.DESCR
    
    else: 
        
        return pd.DataFrame(data = data_all, columns = names_all)  
```

## ***Example: Does it matter when you scale?***


```python
# X_ss = s.fit_transform(X)
# y = california_data["MEDV"]
X_train, X_test, y_train, y_test = train_test_split(X_ss, y, test_size = 0.3, random_state = 72018)
```

```python
# creamos un modelo de regresión lineal
lr = LinearRegression()

# entrenamos nuestro modelo de regresion lineal usando la data X_train, y_train
lr.fit(X_train, y_train)

# predecimos un conjunto de valores con nuestro modelo de regresion lineal usando la data X_test
y_pred = lr.predict(X_test)

# calculamos R^2 teniendo en cuenta el conjunto de datos de prueba y la data predecida y_pred
r2_score(y_pred, y_test)
```

```python
# X = california_data.drop("MEDV", axis=1)
# y = california_data["MEDV"]
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state = 72018)
```

```python
# creamos un modelo de escala estandar
s = StandardScaler()

# creamos un modelo de regresion lineal
lr_s = LinearRegression()

# entrenamos el modelo de escala estandar con la data X_train
X_train_s = s.fit_transform(X_train)

# entrenamos el modelo de regresion lineal con  la data X_train_s, y_train
lr_s.fit(X_train_s, y_train)

# creamos un conjunto de datos con el modelo de escala estandar y la data X_test
X_test_s = s.transform(X_test)

# predecimos un conjunto de datos con nuestro modelo de regresion lineal usando la data X_test_s
y_pred_s = lr_s.predict(X_test_s)

# calculamos R^2 teniendo en cuenta el conjunto de datos de prueba y la data predecida y_pred_s
r2_score(y_pred_s, y_test)
```

**Conclusion:** It doesn't matter whether you scale before or afterwards, in terms of the raw predictions, for Linear Regression. However, it matters for other algorithms. Plus, as we'll see later, we can make scaling part of a `Pipeline`.
